### access pre-trained models with MDL class

We exposed a wide range of APIs to let you query and download our models.
These APIs basically can be accessed via any HTTP requests.
For convenience, we implemented some of the most-used APIs and wrapped them into XenonPy.
All these functions can be accessed using `xenonpy.mdl.MDL`.

```python
# --- import necessary libraries

from xenonpy.mdl import MDL
```

```python
# --- init and check

mdl = MDL()
mdl

mdl.version
```

Noticed that ``mdl`` has optional parameter ``api_key`` and ``endpoint``.
``endpoint`` is as it's name point where to fetch data. ``api_key`` is a access token used to validate the authorization and action permissions.
At this moment, the defaulat key, ``anonymous.user.key`` is the only acceptable option.

### querying

There are many ways to query models. The most straightforward method is use `mdl` method.
It accept variable keywords as input and any keyword hitted will be returned. For example, query models can predict property **refractive index**:

```python
# --- query data

query = mdl('refractive')
query
```

You can see that run a querying method dose not execute the querying immediately but return a queryable object.
If you print out the object, the `queryable` list will be shown. Only the variables in the list can be fetched from server.

Another way to the `queryable` list is call `query.queryable` as below:

```python
query.queryable
```

Let's say way only want to know the variables of `modelset`, `method`, `property`, `descriptor`, `meanAbsError`, `meanSquareError`, `pValue`, and `pearsonCorr`. Execute the query like this:

```python
query(
    'modelset',
    'method',
    'property',
    'descriptor',
    'meanAbsError',  
    'meanSquareError',
    'pValue',
    'pearsonCorr' 
)
```

If everything goes right, you will got the returns in pandas DataFrame.
You can see that 3599 models matched the keyword and these models are contained in two modelset.
It's noticed that all variables in column `pValue` are `None`. This is not a querying error, all variables are `None` able.

BTW, you can retrieve the last querying result from the query object via the `results` property.

```python
query.results.head(3)
```

Querying with `mdl` is simple but not efficient enough. In most cases, we exactly know want we want.

Let's say we want to retrieve some models that trained in the inorganic modelset and can predict the property of refractive index.
In this case, we need to feed the parameter `modelset_has` with **inorganic** and the `property_has` with **refractive**, respectively.

```python
# --- query data

mdl(modelset_has='inorganic', property_has='refractive')()
```

You can see that only the models in the **All inorganic compounds in materials project** modelset were returned. Also, if you call a query object without parameters, all queryable variables will return.

### get training info/env and download url by model id

Maybe you have been noticed that all models have their unique ids. We can use model id to get more information about a model. The following show how to get training info/env by model id.

```python
info = mdl.get_training_info(model_id=1234)()
info.tail(3)
info.plot(y=['train_mse_loss', 'val_mse'])
```

```python
mdl.get_training_env(model_id=1234)()
```

Get download url can be done in a similar way.

```python
mdl.get_model_urls(1234, 5678)()
```

See the column named `url`. Maybe you only want to get the string of `url`, just use `queryable` specification.

```python
mdl.get_model_urls(1234, 5678)('url')
```

Also, if you don't need to return a dataframe or you want to control the output type yourself, you can set `return_json` to `True`.

```python
mdl.get_model_urls(1234, 5678)('url', return_json=True)
```

You can use these urls to download models yourself but we suggest that you to use `mdl.pull`.

```python
mdl.pull?
```

```python
ret = mdl.pull(1234, 5678)
ret
```

column named `model` contains the local path where downloaded models are.

### upload model

Upload models is not availiable until we open the public registration. If you try to upload models, you will get `operation needs a logged in user and the corresponded api_key` error.

```python
mdl.upload_model(
    modelset_id=2, 
    describe=dict(
        property='test2',
        descriptor='test2',
        method='test',
        lang='test',
        deprecated=True,
        isRegression=True,
        meanAbsError=1.11,
        pearsonCorr=0.85,
    ),
    training_info=dict(a=1, b=2),
    supplementary=dict(true=[1,2,3,4], pred=[1,2,2,4])
)(file='set1')
```

### retrieve model

You can use `xenonpy.model.training.Checker` to reload the downloaded models. For example, try to load the model with id **1234**

```python
from xenonpy.model.training import Checker
```

```python
checker = Checker(ret[ret.id == 1234].model.item())
checker
```

something like *$WEnkZ6e3* under filename is a magic number to guarantee that each model has a unique name.

To load model into python, call `checker.model` property.

```python
checker.model
```

Use `checker.checkpoints` to list checkpoints.

```python
checker.checkpoints
```

and `checker.checkpoints[<checkpoint name>]` to load checkpoint.

```python
checker.checkpoints['mse_3']
```

### resue model using trainer

`xenonpy.model.training.Trainer` can load model and checkpoints from checker all from model dir directly.

```python
from xenonpy.model.training import Trainer
```

```python
Trainer.load?
```

```python
trainer = Trainer.load(from_=checker)
trainer
```

The following codes show how to reuse model as predictor.

```python
# if you have not have the samples data
# preset.build('mp_samples', api_key=<your materials project api key>)

from xenonpy.datatools import preset

data = preset.mp_samples
data.head(3)
```

```python
from xenonpy.descriptor import Compositions

prop = data['efermi'].dropna().to_frame()  # reshape to 2-D
desc = Compositions(featurizers='classic').transform(data.loc[prop.index]['composition'])

desc.head(3)
prop.head(3)
```

```python
y_pred = trainer.predict(x_in=torch.tensor(desc.values, dtype=torch.float)).detach().numpy().flatten()
y_true = prop.values.flatten()

draw(y_true, y_pred, prop_name='Efermi ($eV$)')
```
